# El mapa logistico
Ecuacion logistica: $ x_{t +1}=x_{t} \times r(1-x_{t})$ 

Caos: osilaciones a periodicas que no siguien un patron


 


```python
# Title: Logistic map
# Author: Sara Brborich
# Date Created: 30/03/2023
# Date Modified: 30/03/2023
# Description:  class logisticMap para la trayectoria del mapa logistico

import matplotlib.pyplot as plt 
import numpy as np

class logisticMap:
    
    def __init__(self, r, num_iterations, x0): # r es la tasa de crecimiento y num_iterations es el numero de iteraciones, x0 es el valor inicial
        self.r = r
        self.num_iterations = num_iterations
        self.x = [x0] 
        for i in range(self.num_iterations): 
            self.x.append(self.r * self.x[i] * (1 - self.x[i])) 

    def __str__(self):
        modelSummary= (f"********************"
                        f"\nLogistic Map Summary:"
                        f"\nr = {self.r}" 
                        f"\nNumber of iterations = {self.num_iterations}"
                        f"\nInitial value = {self.x[0]}"                        
                        f"\nFinal value = {self.x[-1]}" 
                        f"\n********************")
        return modelSummary

    def max_xt_value(self):
        return max(self.x)
    
    def last_xt_value(self):
        return self.x[-1]

    def xt_steady_state(self):
        return (1 - (1/self.r))

    def plot_xt_evolution(self, showFig):
        fig, ax = plt.subplots()
        ax.plot(self.x, 'o-', label=f'xt') 
        ax.set_ylabel('xt')
        ax.set_xlabel('Time')
        ax.legend()
        if showFig:
            plt.show()
                
        return fig   
    
      
```

```python
model1 = logisticMap(3.5, 100, 0.1) 
print(model1) 

model1.plot_xt_evolution(True)
```

```python
## usar linespace para generar 1000 valores de r entre 0 y 4
r = np.linspace(2.5, 4, 10000)

## crear una lista vacia para guardar los valores de xt
xt = []

for i in r:
    model = logisticMap(i, 1000, 0.1)
    xt.append(model.last_xt_value())



fig, ax = plt.subplots()
ax.plot(r, xt, 'o', label=f'xt', markersize=0.4) 
ax.set_ylabel('xt')
ax.set_xlabel('r')
ax.legend()

plt.show()
    
```
